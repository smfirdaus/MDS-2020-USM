## Explore dataset A

```python
# clustermap heat map
figA = plt.gcf()
figA.set_size_inches(20, 20)
sns.heatmap(np.around(dfA.corr(),2),square=True,annot=True,linewidths=.2)
plt.show()
```

```python
profA = ProfileReport(dfA)
profA.to_file(output_file='profileMeterA.html')
```

## Explore dataset C

```python
# clustermap heat map
figC = plt.gcf()
figC.set_size_inches(20, 20)
matrix = np.triu(dfC.corr()) # to mask the upper shape of the correlation matrix
sns.heatmap(np.around(dfC.corr(),2),square=True,annot=True,linewidths=.2, mask=matrix)
plt.savefig('heatdfC.png')
plt.show()
```

Comment: Generally, if the correlation between a pair of variables is greater than 0.5-0.6, we should seriously consider dropping one of those variables

```python
# run input-target correlation 
dfCcor = dfC.corr()
dfCcor = dfCcor.sort_values(by=['class'])
dfCcor[1:43].plot(y='class', kind='barh', figsize=(8,15), legend=False)
plt.savefig('corrdfC.png')
plt.show()
```

```python
profC = ProfileReport(dfC)
profC.to_file(output_file='profMeterC.html')
```
